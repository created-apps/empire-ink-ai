## Theory 1 — Variables & Data Types

A **variable** is a named box that stores a value. Python has several basic data types:

| Type | Example | Use |
|------|---------|-----|
| `str` | `"Akbar"` | Text |
| `int` | `1556` | Whole numbers |
| `float` | `3.14` | Decimal numbers |
| `bool` | `True` | Yes/No flags |

```python
title = "Baburnama Folio"
era = "Akbari"
year_created = 1598
is_illustrated = True
scale_cm = 23.5

print(f"Title: {title}")
print(f"Era: {era}")
print(f"Year: {year_created}")
print(f"Illustrated: {is_illustrated}")
print(f"Scale: {scale_cm} cm")
```

## Theory 2 — Functions & f-strings

A **function** is a reusable block of code. You define it once with `def` and call it whenever you need it. **f-strings** embed variables directly inside strings using `{}`.

```python
def greet_artist(name):
    return f"Welcome to Empire & Ink, {name}!"

message = greet_artist("Basawan")
print(message)

def describe_painting(title, artist, year):
    return f"'{title}' by {artist}, circa {year}."

desc = describe_painting("Hamzanama", "Mir Sayyid Ali", 1562)
print(desc)
```

## Theory 3 — Dictionaries & Loops

A **dictionary** stores key-value pairs — perfect for structured data like a painting record. A **for loop** repeats code for each item in a collection.

```python
painting = {
    "title": "Akbarnama Folio",
    "artist": "Basawan",
    "era": "Akbari",
    "year": 1590,
    "medium": "opaque watercolour on paper",
}
print(painting["title"])
print(painting.get("artist", "Unknown"))

gallery = [
    {"title": "Baburnama", "era": "Akbari", "year": 1598},
    {"title": "Padshahnama", "era": "Shahjahani", "year": 1656},
    {"title": "Tuzuk-i-Jahangiri", "era": "Jahangiri", "year": 1615},
]
for item in gallery:
    print(f"{item['year']} — {item['title']} ({item['era']})")
```
