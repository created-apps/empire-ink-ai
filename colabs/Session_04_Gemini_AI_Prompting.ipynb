## Theory 1 — What is an LLM & How Image Generation Works

An **LLM** (Large Language Model) like Gemini is trained on billions of text examples. It predicts the most likely next tokens — the result feels like understanding because the patterns are so rich.

**Image generation** works differently: start with random pixels, nudge them towards a prompt description using a diffusion model, repeat thousands of times — and a coherent image emerges.

**Our pipeline:**
```
User prompt → Gemini Flash (text) → Enhanced Mughal prompt → Imagen 3 (image) → PNG file
```

````python
def simulate_pipeline(user_prompt):
    enhanced = (
        f"[Mughal miniature] {user_prompt}, akbari period, "
        "opaque watercolour on paper, intricate border, gold leaf"
    )
    return enhanced, f"[Image from: {enhanced[:50]}...]"

enhanced, image = simulate_pipeline("A tiger in a garden")
print("Enhanced:", enhanced)
print("Image:   ", image)
````

## Theory 2 — Prompt Engineering Principles

Good prompts are **specific, styled, and structured**.

| Weak | Strong |
|------|--------|
| "a bird" | "White falcon on a jewelled glove, Akbari Mughal miniature, opaque watercolour on wasli paper, intricate floral border in gold and lapis lazuli, circa 1590" |

**Mughal style descriptors:**
- **Akbari** (1556–1605): bold lines, Persian influence, vibrant reds and greens, crowded compositions
- **Jahangiri** (1605–1627): naturalism, detailed flora and fauna, soft muted colours
- **Shahjahani** (1628–1658): refined elegance, pastel palette, architectural marble settings

````python
STYLE_DESCRIPTORS = {
    "akbari": (
        "Akbari period Mughal miniature, bold Persian-influenced outlines, "
        "vibrant vermillion and lapis lazuli palette, crowded multi-figure composition, "
        "opaque watercolour on wasli paper, intricate geometric border, circa 1580"
    ),
    "jahangiri": (
        "Jahangiri period Mughal miniature, naturalistic flora and fauna, "
        "European chiaroscuro influence, soft muted palette with gold highlights, "
        "delicate brushwork, opaque watercolour, circa 1615"
    ),
    "shahjahani": (
        "Shahjahani period Mughal miniature, refined courtly elegance, pastel palette, "
        "architectural marble settings with pietra dura inlay, floral border in gold leaf, "
        "symmetrical composition, opaque watercolour on fine wasli paper, circa 1645"
    ),
}
for style, desc in STYLE_DESCRIPTORS.items():
    print(f"{style.upper()}: {desc[:80]}...")
````

### Challenge 4 — batch_generate()

````python
import time

def batch_generate(prompts):
    # YOUR CODE HERE — prompts: list of {"prompt": str, "style": str}
    # Returns list of {"prompt", "style", "enhanced", "status"}
    # Add 2s delay between calls to respect rate limits
    results = []
    for item in prompts:
        try:
            enhanced = enhance_prompt(item["prompt"], item.get("style", "akbari"))
            results.append({"prompt": item["prompt"], "style": item.get("style"),
                           "enhanced": enhanced, "status": "ok"})
        except Exception as e:
            results.append({"prompt": item["prompt"], "error": str(e), "status": "error"})
        time.sleep(2)
    return results

batch = [{"prompt": "a tiger", "style": "akbari"}, {"prompt": "a peacock", "style": "jahangiri"}]
for r in batch_generate(batch):
    print(r["status"], "—", r["prompt"])
````
